```python
import requests
from bs4 import BeautifulSoup
import re
# import lxml
```

```python
url='https://www.bechtle.com/impressum'
```

```python
scraped_url=requests.get(url)
soup=BeautifulSoup(scraped_url.content,'html.parser')
body=soup.find('body').text
body=body.replace(u'\xa0',u' ')
# # html=html.replace(u'\n',u' ')
body_list=body.split("\n")
```

#email with reg ex!!

```python
# Make a regular expression for validating an Email

email_regex = r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b'
 
# Define a function for validating an Email
 
def check_email(email):

    # pass the regular expression and the string into the fullmatch() method

    if(re.search(email_regex, email)):
        print(email)
```

```python

for add_element in body_list:
  check_email(add_element)
```
